```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

## General Description of Data

```python
df.describe()
```

## Gender_wise Count Of Students

```python
mpt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = 'gender')
ax.bar_label(ax.containers[0])
mpt.title("Gender-wise Count of Students", size = 18)
```

Above data cleary shows that there are more female students than the male students.

```python
gb = df.groupby('parental level of education').agg({'math score':'mean','reading score':'mean', 'writing score':'mean'})
```

```python
mpt.figure(figsize = (5,5))
ax = sns.heatmap(gb, annot = True)

mpt.title("Distribution of Marks over Parent's Education", size = 18)
```

Marks of students whose parents are more educated are higher as compare to the marks of students whose parents are less
educated and also math score are comparatively less that that of reading and writing score.

```python
gb1 = df.groupby('test preparation course').agg({'math score':'mean','reading score':'mean', 'writing score':'mean'})
```

```python
sns.heatmap(gb1, annot = True)
mpt.title("Course-wise Marks", size = 18)
```

Marks of students who have completed the test preparation course are comparatively higher than the marks of students who haven't completed the test.

```python
groupA = df.loc[(df['race/ethnicity'] == "group A")].count()
groupB = df.loc[(df['race/ethnicity'] == "group B")].count()
groupC = df.loc[(df['race/ethnicity'] == "group C")].count()
groupD = df.loc[(df['race/ethnicity'] == "group D")].count()
groupE = df.loc[(df['race/ethnicity'] == "group E")].count()
```

```python
my_list = [groupA['race/ethnicity'], groupB['race/ethnicity'], groupC['race/ethnicity'], groupD['race/ethnicity'], groupE['race/ethnicity']]

l = ["groupA","groupB","groupC","groupD","groupE"]

mpt.pie(my_list,autopct = "%1.2f%%", labels = l)
mpt.title("Race/Ethnicity Distribution", size = 18)

```

In the above pie-chart it is shown that most of the students belongs to groupC and there are only few students who belongs to groupA

### Conclusion:

As of the analysis we can successfully conclude that the number of female students are higher than the number of male 
students and score of students whose parent's are well-educated have good marks. The students that have taken and
completed the test preperation course have scored higher marks than those students who haven't completed the test. Majority
of students belongs to group C ethnicity/race.
